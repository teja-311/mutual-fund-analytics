## Insight 1: Downside Risk Differs Across Funds

Historical VaR and CVaR analysis revealed meaningful differences in downside risk across mutual funds.

Funds with higher return potential generally exhibited larger tail losses, highlighting the tradeoff between risk and return.

## Insight 2: Risk-Adjusted Performance Changes Over Time

Rolling 90-day Sharpe ratios varied significantly through time.

Periods of market strength improved risk-adjusted returns, while volatile periods caused Sharpe ratios to decline.

## Insight 3: Investor Behavior Varies by Cohort

Cohort analysis showed that investors entering in different years demonstrated different investment patterns and fund preferences.

Certain cohorts contributed disproportionately to total invested capital.

## Insight 4: Long-Term SIP Investors Form a Core Base

SIP continuity analysis identified a group of investors who maintained systematic investments consistently over time.

These investors contribute significantly to stable fund inflows.

## Insight 5: Portfolio Diversification Varies Across Funds

HHI concentration scores indicated that some funds maintain highly diversified portfolios while others concentrate investments in fewer holdings.

Higher concentration may increase potential returns but also increases portfolio-specific risk.
